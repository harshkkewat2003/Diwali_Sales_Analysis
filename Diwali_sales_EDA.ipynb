#Data cleaning  :-

```python
#drop/delete blank columns

df.drop(['Status', 'unnamed1'], axis=1, inplace=True)
```

```python
#check for null values

pd.isnull(df).sum()
```

```python
#drop null values

df.dropna(inplace=True)
```

```python
#change data types

df['Amount'] = df['Amount'].astype('int')
```

#Data Manipulation

```python
# returns Age greater than 18

df[df['Age']>18]
print(df)
```

```python
# returns Age less than 60

df[df['Age']>50]
```

```python
# sort largest to lowest  amount shopping

df.sort_values('Amount')
```

```python
#returns 5 largest amount sales

df.nlargest(5,'Amount')
```

```python
#renaming columns values

df['Marital_Status'] = df['Marital_Status'].replace({ 0:'Unmarried',1: 'Married' })
```

#Exploratory Data Analysis

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

1. Gender

```python

# Gender (Categorical data) - Countplot


sns.countplot(data=df, x="Gender")
plt.title("Gender Distribution", fontsize=16)
plt.xlabel("Gender", fontsize=14)
plt.ylabel("Count", fontsize=14)
plt.show()
```

From above graphs we can see that most of the buyers are females .

#2. Age

```python
# Age Group (Categorical data) - Countplot


sns.countplot(data=df, x="Age Group")
plt.title("Age Group Distribution", fontsize=16)
plt.xlabel("Age Group", fontsize=14)
plt.ylabel("Count", fontsize=14)
plt.show()
```

```python
#Which age group has done highest no. of shoppings

age_counts = df['Age Group'].value_counts()

age_counts.plot(kind='pie', autopct='%1.1f%%')
plt.title('Age Group Distribution')
plt.ylabel('')
plt.show()
```

From above graphs we can see that most of the buyers are of age group between 26-35 yrs female

#3.  Maritial Status

```python
# Marital_Status (Categorical data) - Countplot


sns.countplot(data=df, x="Marital_Status")
plt.title("Marital Status Distribution", fontsize=16)
plt.xlabel("Marital Status", fontsize=14)
plt.ylabel("Count", fontsize=14)
plt.show()
```

From above graphs we can see that most of the buyers are unmarried and they have high purchasing power

#4.  State

```python
# State (Categorical data) - Countplot

plt.figure(figsize=(12, 6))
sns.countplot(data=df, x="State")
plt.title("State Distribution", fontsize=16)
plt.xlabel("State", fontsize=14)
plt.ylabel("Count", fontsize=14)
plt.xticks(rotation=90)
plt.show()
```

From above graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh and Maharashtra.

#5.  Occupation

```python
# Occupation (Categorical data) - Countplot


sns.countplot(data=df, x="Occupation")
plt.title("Occupation Distribution", fontsize=16)
plt.xlabel("Occupation", fontsize=14)
plt.ylabel("Count", fontsize=14)
plt.xticks(rotation=90)
plt.show()
```

From above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation sector

#6. Product Category  :-

```python
sns.countplot(data=df, x="Product_Category")
plt.title("Product Distribution", fontsize=16)
plt.xlabel("Product Category", fontsize=14)
plt.ylabel("Count", fontsize=14)
plt.xticks(rotation=90)
plt.show()
```

From above graphs we can see that most of the sold products are from Food, Clothing and Electronics category

Married women age group 26-35 yrs from UPand Maharastra  working in IT, Healthcare and Aviation are more likely to buy products from Food, Clothing and Electronics category
